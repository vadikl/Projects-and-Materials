````python
import requests
from bs4 import BeautifulSoup
````

### Beautiful Soup Practice
- Parent: The main tags
- Children: Tags embedded into parent tags
- Descendants: Tags embedded into children tags 

### Main Methods
- ```soup.find(tag, class_ = tag_id)``` - finds only first tag with the corresponding id
- ```soup.find_all(tag, class_ = tag_id)``` - finds all elements

````python
# Quotes Extraction

# HTML Getting
url = 'https://quotes.toscrape.com/'
response = requests.get(url).text
page = BeautifulSoup(response, 'lxml')


'''
All Quotes are placed in <span> tags that have class: text. However, by default 
an html object will be returned. To retrive the text, use .text method
'''
# Target Elements Finding
quotes = page.find_all('span', class_= 'text') 
authors = page.find_all('small', class_= 'author')
tags = page.find_all('div', class_= 'tags')

# Iterate Over Them
for idx in range(0, len(quotes)):
    print(quotes[idx].text)
    print(f'Author: {authors[idx].text}')
    
    tags_for_qoute = tags[idx].find_all('a', class_ = 'tag')
    all_tags = ''
    for tag in tags_for_qoute:
        all_tags = all_tags + ' ' + tag.text  
    print(f'Tags: {all_tags}')
    print('\n')
````

````python
# Items Extraction

page_start = 1
page_end = 10
home_page = 'https://scrapingclub.com'

for page in range(page_start, page_end+1):
    current_page = f'https://scrapingclub.com/exercise/list_basic/?page={page}'
    response = requests.get(current_page).text
    page = BeautifulSoup(response, 'lxml')
    div_items = page.find_all('div', class_ = 'col-lg-4 col-md-6 mb-4')

    for idx in range(0, len(div_items)):
        tag_link = div_items[idx].find('a').get('href')
        item_link = f'{home_page}{tag_link}'
        item_response = requests.get(item_link).text
        item_page = BeautifulSoup(item_response, 'lxml')

        item_block = item_page.find('div', class_ = 'card-body')

        item_name = item_block.find('h3', class_ = 'card-title').text
        item_price = item_block.find('h4').text
        item_desc = item_block.find('p').text

        print(f'Name: {item_name}'+'\n'+ 
              f'Price: {item_price}'+'\n'+
              'Description:'+'\n'+item_desc+'\n')  
````
